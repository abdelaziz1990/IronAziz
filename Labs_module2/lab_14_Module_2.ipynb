```python
projects['goal_met']=np.where(projects['usd_pledged_real']>projects['usd_goal_real'],1,0)

projects.head()
```

```python
projects['usd']=np.where(projects['currency']=='USD',1,0)
projects.head()
```

```python
crosstab=pd.crosstab(projects['goal_met'],projects['usd'])
crosstab
```

```python
from scipy.stats import fisher_exact

oddsratio, pvalue = fisher_exact(crosstab)

print(pvalue)
```
