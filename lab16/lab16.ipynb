```python
#Find a Python list with the top ten FBI's Most Wanted names.
html_3 = requests.get('https://www.fbi.gov/wanted/topten').content
html_3


```

```python
#Display the 20 latest earthquakes info (date, cellspacing='0', width='100%'time, latitude, longitude and region name) by the EMSC as a pandas dataframe.
html_4 = requests.get('https://www.emsc-csem.org/Earthquake/').content
earth = BeautifulSoup(html_4, 'html.parser')
table = earth.find_all('table', border='0', cellspacing='0', width='100%')[0]
table
rows = table.find_all('tr')
rows = [row.text.strip().split("\n") for row in rows]
rows
colnames = rows[0]
data = rows[1:]

df = pd.DataFrame(data, columns=colnames)
df





```
