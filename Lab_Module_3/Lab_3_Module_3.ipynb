```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sb
sb.pairplot(iris.dropna(), hue = 'class')
```

```python
iris.loc[iris['class'] == 'versicolor', 'class'] = 'Iris-versicolor'
iris.loc[iris['class'] == 'Iris-setossa', 'class'] = 'Iris-setosa'

iris['class'].unique()
```

```python
iris = iris.loc[(iris['class'] != 'Iris-setosa') | (iris['sepal_width_cm'] >= 2.5)]
iris.loc[iris['class'] == 'Iris-setosa', 'sepal_width_cm'].hist()
```

```python
iris.loc[(iris['class'] == 'Iris-versicolor') &
              (iris['sepal_length_cm'] < 1.0)]
```

```python
iris.loc[(iris['sepal_length_cm'].isnull()) |
              (iris['sepal_width_cm'].isnull()) |
              (iris['petal_length_cm'].isnull()) |
              (iris['petal_width_cm'].isnull())]
```

```python
average_petal_width = iris.loc[iris['class'] == 'Iris-setosa', 'petal_width_cm'].mean()

iris.loc[(iris['class'] == 'Iris-setosa') &
              (iris['petal_width_cm'].isnull()),
              'petal_width_cm'] = average_petal_width
```

```python
iris.to_excel(r'C:\Users\Mtime\OneDrive\Bureau\iris.xlsx', index=False)

```
